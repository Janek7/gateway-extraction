```python
from PetReader import pet_reader
from PetReader import PetReader
from petreader import labels

gold_key_words = ['either', 'for', 'for each patient for which', 'for the case', 'if',
                  'in case', 'in case of', 'in the case of', 'it can also happen that', 'or', 'otherwise', 'should',
                  'sometimes', 'under certain circumstances', 'whereas']
```

## Doc Investigation

```python
doc_name = 'doc-5.4'
doc_relations = PetReader.relations_dataset.GetRelations(pet_reader.get_document_number(doc_name))
doc_flow_relations, doc_same_gateway_relations = doc_relations[labels.FLOW], doc_relations[labels.SAME_GATEWAY]

def filter_flow_relations(flow_relations, entity_type_list):
    """
    filter list of flow relations (single dictionaries) for source or target entity type = given entity tyoe
    """
    return [flow_relation for flow_relation in flow_relations if flow_relation[labels.SOURCE_ENTITY_TYPE] in entity_type_list
                                                               or flow_relation[labels.TARGET_ENTITY_TYPE] in entity_type_list]

doc_flow_relations_xor = filter_flow_relations(doc_flow_relations, [labels.XOR_GATEWAY, labels.CONDITION_SPECIFICATION])
doc_flow_relations_and = filter_flow_relations(doc_flow_relations, [labels.AND_GATEWAY])
doc_flow_relations_gateways = filter_flow_relations(doc_flow_relations, [labels.XOR_GATEWAY, labels.AND_GATEWAY, labels.CONDITION_SPECIFICATION])
print(f"Flow relations involving XOR gateways {len(doc_flow_relations_xor)}; "\
      f"AND gateways {len(doc_flow_relations_and)}; overall gateways {len(doc_flow_relations_gateways)}; overall {len(doc_flow_relations)}")

print("*"*100)

for i, same_gateway_relation in enumerate(doc_same_gateway_relations):
    for key, value in same_gateway_relation.items():
        print(f"{key}: {value}")
    print()

print("*"*100)

for i, flow_relation in enumerate(doc_flow_relations):
    for key, value in flow_relation.items():
        print(f"{key}: {value}")
    print()
```

### Pairs:
detected with manual checks of documents


+ either;in case of
+ sometimes;sometimes (doc-6.4)
+ if;otherwise
+ if;if
+ for;for (only in special cases, e.g. doc-4.1)
+ in case of;otherwise
+ in case;in case
+ in case;otherwise
+ for the case;for the case (doc-2.1)

```python
counter = 0
docs_already_checked = ['doc-6.4', 'doc-6.1', 'doc-3.8', 'doc-5.4', 'doc-3.6', 'doc-1.2', 'doc-10.10', 'doc-10.1', 'doc-2.2',
                       'doc-10.2', 'doc-5.2', 'doc-10.11', 'doc-9.2', 'doc-4.1', 'doc-6.3', 'doc-10.7', 'doc-10.8', 'doc-10.14',
                        'doc-3.3', 'doc-9.5']

for doc_name in pet_reader.document_names:
    doc_text = pet_reader.get_doc_text(doc_name)
    for key_word in gold_key_words:
        if f" {key_word} " in doc_text.lower() and key_word != "for" and doc_name not in docs_already_checked:
            print(f"{doc_name}".center(100, '-'))
            print(key_word)
            counter += 1
            for i, sentence in enumerate(pet_reader.get_doc_sentences(doc_name)):
                key_word_found = "X"
                for key_word2 in gold_key_words:
                    if f" {key_word2} " in f" {sentence.lower()} " and key_word2 != "for":
                        key_word_found = key_word2
                print(i, key_word_found.ljust(10), "|", sentence)
            break
print(f"{counter} out of {len(pet_reader.document_names)}")
```
